```python
# Install ZBar which isrequired for pyzbar to decode QR codes

!apt-get install libzbar0 -y

# install Python packages
!pip install qrcode opencv-python pyzbar

# import libraries which is required for our qr code encoder/decodder
import qrcode
from pyzbar.pyzbar import decode
import cv2
from PIL import Image


# function to generate the QR code
def generate_qr(data, filename="my_qr.png"):
    qr = qrcode.make(data)
    qr.save(filename)
    print(f"QR Code saved as '{filename}'")
    display(Image.open(filename))


# function to decode the qr code
def read_qr(filename):
    img = cv2.imread(filename)
    decoded_objs = decode(img)


    if not decoded_objs:
        print("QR Code not found.")
    else:
        for obj in decoded_objs:

            print("Decoded Data:", obj.data.decode("utf-8"))

print("Fatima QR Code Generator ")

# Choosing options
print("1: Generate QR \n2: Decode QR ")

choice = input("Enter choice (1 or 2): ")


if choice == "1":
    text = input("Enter text or URL to encode: ")
    filename = "generated_qr.png"

    generate_qr(text, filename)


elif choice == "2":
    from google.colab import files

    print(" Please upload your QR Code image.")
    uploaded = files.upload()


    for fn in uploaded.keys():
        read_qr(fn)

else:
    print("Invalid choice.")

```
