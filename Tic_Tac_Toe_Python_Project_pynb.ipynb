```python
# fun tic tac toe game in python

# board with 9 empty spaces
board = [" " for i in range(9)]

# function to show the board
def print_board():
    print()
    for i in range(3):
        print(board[i*3], "|", board[i*3+1], "|", board[i*3+2])
        if i < 2:
            print("--+---+--")
    print()

# function to check if a player wins
def check_win(player):
    # all winning combinations
    win_combo = [
        [0,1,2], [3,4,5], [6,7,8], # rows
        [0,3,6], [1,4,7], [2,5,8], # columns
        [0,4,8], [2,4,6]           # diagonials
    ]

    for combo in win_combo:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] == player:
            return True
    return False

# check if game is draw when no spaces left
def check_draw():
    return " " not in board

# main game class
class TicTacToe:
    def __init__(self):
        self.current_player = "X"

    def play(self):
        print("Welcome to Fatima's Tic-Tac-Toe Game üòé‚ùå‚≠ï")
        while True:
            print_board()
            print("Player", self.current_player, "'s turn")

            try:
                move = int(input("Enter position (1-9): ")) - 1

                if move < 0 or move > 8:
                    print("Please choose a number from 1 to 9.")
                    continue

                if board[move] != " ":
                    print("This place is already taken. Try again.")
                    continue

                board[move] = self.current_player

                if check_win(self.current_player):
                    print_board()
                    print("Player", self.current_player, "won the game! üéâ")
                    break


                if check_draw():
                    print_board()
                    print("It's a draw! üòÖüòÖ")
                    break

                # switch player
                self.current_player = "O" if self.current_player == "X" else "X"

            except ValueError: # short error handling
                print("Invalid input. Only numbers allowed.")

# Ruun the game
game = TicTacToe()
game.play()
```
